```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot z, v, a
def plot(data):
    # Calculate the z-acceleration
    data['az'] = np.gradient(data['vz'], data['t'])

    # Create a figure with three subplots stacked vertically
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 6), sharex=True)

    # Plot z-position in the first subplot
    ax1.plot(data['t'], data['z'], 'b-', linewidth=1.5)
    ax1.set_ylabel('Z-Position')
    ax1.grid(True)

    # Plot z-velocity in the second subplot
    ax2.plot(data['t'], data['vz'], 'r-', linewidth=1.5)
    ax2.set_ylabel('Z-Velocity')
    ax2.grid(True)

    # Plot z-acceleration in the third subplot
    ax3.plot(data['t'], data['az'], 'g-', linewidth=1.5)
    ax3.set_xlabel('Time')
    ax3.set_ylabel('Z-Acceleration')
    ax3.grid(True)

    # set time range
    #plt.xlim(0, 20)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```

```python
# plot compression and force of left spring

def plotSpring(data):
    # Create a figure with three subplots stacked vertically
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4), sharex=True)

    # Plot compressoin in the first subplot
    ax1.plot(data['t'], data['dl_left'], 'b-', linewidth=1.5)
    ax1.set_ylabel('compression left spring')
    ax1.grid(True)

    # Plot force in the second subplot
    ax2.plot(data['t'], data['F_left'], 'r-', linewidth=1.5)
    ax2.set_ylabel('F_left')
    ax2.grid(True)

    # set time range
    #plt.xlim(0, 20)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```

```python
# plot bumper z, v, a
def plotBumper(data):
    # Calculate the z-acceleration
    data['az_bumper'] = np.gradient(data['vz_bumper'], data['t'])

    # Create a figure with three subplots stacked vertically
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 6), sharex=True)

    # Plot z-position in the first subplot
    ax1.plot(data['t'], data['z_bumper'], 'b-', linewidth=1.5)
    ax1.set_ylabel('Z-Position Bumper')
    ax1.grid(True)

    # Plot z-velocity in the second subplot
    ax2.plot(data['t'], data['vz_bumper'], 'r-', linewidth=1.5)
    ax2.set_ylabel('Z-Velocity Bumper')
    ax2.grid(True)

    # Plot z-acceleration in the third subplot
    ax3.plot(data['t'], data['az_bumper'], 'g-', linewidth=1.5)
    ax3.set_xlabel('Time')
    ax3.set_ylabel('Z-Acceleration Bumper')
    ax3.grid(True)

    # set time range
    #plt.xlim(0, 20)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```

```python
# plot Impulsbilanz

def plotImpulsbilanz(data):

    m_car = 0.4 # kg
    m_bumper = 0.4 # kg
    
    # Calculate the z-momentum of the car
    data['p_z'] = m_car * data['vz']

    # Calculate the z-momentum of the bumper
    data['p_z_bumper'] = m_bumper * data['vz_bumper']


    # Create a figure with three subplots stacked vertically
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4), sharex=True)
    fig.suptitle('Impulsbilanz')

    # Plot individual momentums in the first subplot
    ax1.plot(data['t'], data['p_z'], 'b-', linewidth=1.5)
    ax1.plot(data['t'], data['p_z_bumper'], 'r-', linewidth=1.5)
    ax1.set_ylabel('p_z')
    ax1.grid(True)

    # Plot total momentum
    data['p_z_total'] = data['p_z'] + data['p_z_bumper']
    ax2.plot(data['t'], abs(data['p_z_total']), 'g-', linewidth=1.5)
    ax2.set_ylabel('|p_z_total|')
    ax2.grid(True)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```

```python
# plot Energiebilanz

def plotEnergiebilanz(data):

    m_car = 0.4 # kg
    m_bumper = 0.4 # kg

    # Calculate the kinetic energy of the car
    data['E_kin_car'] = 0.5 * m_car * data['vz']**2

    # Calculate the kinetic energy of the bumper
    data['E_kin_bumper'] = 0.5 * m_bumper * data['vz_bumper']**2
    
    
    # Create a figure with three subplots stacked vertically
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 4), sharex=True)
    fig.suptitle('Energiebilanz')

    # Plot individual kinematic energies
    ax1.plot(data['t'], data['E_kin_car'], 'b-', linewidth=1.5)
    ax1.plot(data['t'], data['E_kin_bumper'], 'r-', linewidth=1.5)
    ax1.set_ylabel('E_kin')
    ax1.grid(True)

    # Plot total energy
    data['E_kin_total'] = data['E_kin_car'] + data['E_kin_bumper']
    ax2.plot(data['t'], data['E_kin_total'], 'g-', linewidth=1.5)
    ax2.set_ylabel('E_kin_total')
    ax2.grid(True)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()
```

```python
file_path = 'time_series vinit1 kf10 20ms.csv'
data = pd.read_csv(file_path, delimiter=',')

plot(data)
plotSpring(data)
```
